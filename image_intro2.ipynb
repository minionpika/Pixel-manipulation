```python
# -*- coding: utf-8 -*-
"""

Original template:
https://cs.brown.edu/courses/csci1430/2021_Spring/resources/python_tutorial/
Edited by
@author: murali subbarao, sbu, ece
"""

import numpy as np
from skimage import io, color, data
import matplotlib.pyplot as plt
from PIL import Image

print('--------------------------------------1---------------------------------------------')

# Read in original RGB image.
A = io.imread('https://drive.google.com/uc?export=view&id=1N2OE27M7aurNFc8BW2PzRYFro0vyhsjs')
(m,n,o) = A.shape
plt.imshow(A)
plt.show()


print('--------------------------------------2---------------------------------------------')

# Extract color channels.
RC = A[:,:,0] # Red channel
GC = A[:,:,1] # Green channel
BC = A[:,:,2] # Blue channel

# Create an all black channel.
allBlack = np.zeros((m, n), dtype=np.uint8)

# Create color versions of the individual color channels.Display and save them individually.
justRed = np.stack((RC, allBlack, allBlack), axis=2)
justGreen = np.stack((allBlack, GC, allBlack), axis=2)
justBlue = np.stack((allBlack, allBlack, BC), axis=2)

plt.imshow(justRed)
plt.show()
plt.imshow(justGreen)
plt.show()
plt.imshow(justBlue)
plt.show()
io.imsave('/content/drive/MyDrive/Colab Notebooks/assignment1/images/justRed.jpeg' , justRed)
io.imsave('/content/drive/MyDrive/Colab Notebooks/assignment1/images/justGreen.jpeg' , justGreen)
io.imsave('/content/drive/MyDrive/Colab Notebooks/assignment1/images/justBlue.jpeg' , justBlue)


print('--------------------------------------3---------------------------------------------')

# Compute grey level image from avg. of three channels.
AG = np.mean(A, axis=2)
AG = AG.astype(np.uint8)
plt.imshow(AG, cmap=plt.cm.gray)
plt.show()


print('--------------------------------------4---------------------------------------------')

#compute the histogram
redcHist=np.zeros(256, dtype = np.intc)
greencHist=np.zeros(256, dtype = np.intc)
bluecHist=np.zeros(256, dtype = np.intc)
greycHist=np.zeros(256, dtype = np.intc)

for i in range(m) :
    for j in range(n) :
        redcHist[justRed[i,j,0]] += 1
        greencHist[justGreen[i,j,1]] += 1
        bluecHist[justBlue[i,j,2]] += 1
        greycHist[AG[i,j]] += 1


#plot the histogram
plt.plot(redcHist)
plt.show()
plt.plot(greencHist)
plt.show()
plt.plot(bluecHist)
plt.show()
plt.plot(greycHist)
plt.show()


print('--------------------------------------5---------------------------------------------')


thresholdBrightness = int(input("threshold brightness = "))
grayArray = np.array(AG)
binaryArray = np.zeros((m, n), dtype=np.uint8)

for y in range(n):
    for x in range(m):
        if AG[x, y] >= thresholdBrightness:
            binaryArray[x, y] = 255
        else:
            binaryArray[x, y] = 0

plt.imshow(binaryArray, cmap=plt.cm.gray)
plt.show()


print('--------------------------------------6---------------------------------------------')


thresholdEdge = int(input("threshold Edge = "))
edgeDetectedArray = np.zeros((m, n), dtype=np.uint8)

Gx = np.zeros((m, n), dtype=np.uint8)
for y in range(0, m-1):
    Gx[y,:] = AG[y+1, :] - AG[y, :]

Gy = np.zeros((m, n), dtype=np.uint8)
for x in range(0, n-1):
    Gy[:,x] = AG[:, x+1] - AG[:, x]

Gm = np.sqrt(Gx*Gx + Gy*Gy);

for y in range(n):
    for x in range(m):
        if Gm[x, y] > thresholdEdge:
            edgeDetectedArray[x, y] = 255
        else:
            edgeDetectedArray[x, y] = 0

plt.imshow(edgeDetectedArray, cmap=plt.cm.gray)
plt.show()


print('--------------------------------------7---------------------------------------------')


Ag2 = np.zeros((m//2, n//2), dtype=np.uint8)
Ag4 = np.zeros((m//4, n//4), dtype=np.uint8)
Ag8 = np.zeros((m//8, n//8), dtype=np.uint8)

for y in range(n//2):
    for x in range(m//2):
        block = [AG[x*2, y*2], AG[x*2+1, y*2], AG[x*2, y*2+1], AG[x*2+1, y*2+1]]
        Ag2[x,y] = (np.mean(block)).astype(np.uint8)

for y in range(0, n//4):
    for x in range(0, m//4):
        block = [Ag2[x*2, y*2], Ag2[x*2+1, y*2], Ag2[x*2, y*2+1], Ag2[x*2+1, y*2+1]]
        Ag4[x,y] = (np.mean(block)).astype(np.uint8)

for y in range(0, n//8):
    for x in range(0, m//8):
        block = [Ag4[x*2, y*2], Ag4[x*2+1, y*2], Ag4[x*2, y*2+1], Ag4[x*2+1, y*2+1]]
        Ag8[x,y] = (np.mean(block)).astype(np.uint8)

plt.imshow(Ag2, cmap=plt.cm.gray)
plt.show()
plt.imshow(Ag4, cmap=plt.cm.gray)
plt.show()
plt.imshow(Ag8, cmap=plt.cm.gray)
plt.show()

```
